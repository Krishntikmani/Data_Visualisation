#### Line Plots
-> Requrements:
- x-axis points
- y-axis points


-> It draws a line through given points

```python
x = np.arange(10)
print(x)
y1 = x**2
y2 = 2*x + 3
print(y1)
print(y2)
```

```python
#now say suppose we want to plot one x vs y1
plt.plot(x,y1,color='red',label="Apple",marker='o')
plt.legend()
plt.show() #this plt.show() marks the end of one graph
```

```python
#now we want to plot all the lines in one graphs
plt.plot(x,y1,color='red',label='Apple',marker='o')
plt.plot(x,y2,color='green',label='Kiwi',marker='o',linestyle='dashed')
plt.xlabel('Time')
plt.ylabel('Fruits Sold')
plt.title('Fruits sold with time')
plt.legend()
plt.show()
```

```python
#now suppose we give only one array to the plot function
prices = np.array([1,2,3,4])**2
plt.plot(prices)
plt.show()
#now what happens here is that prices are plotted in y axis and x axis values are treated as indices
```

#### Scatter Plots
- it makes a plot of points

```python
#to adjust the plot size we can use the figure method
plt.figure(figsize=(10,5))
plt.scatter(x,y1,color='red',label='Apple',marker='^')
plt.scatter(x,y2,color='green',label='Kiwi',marker='o',linestyle='dashed')
plt.xlabel('Time')
plt.ylabel('Fruits Sold')
plt.title('Fruits sold with time')
plt.legend()
plt.show()
```

#### Bar Graphs
- plots the height of different x-tick labels
- we can manage the width of the bars
- when we start give x-coodinates a array to plot different heights

```python
plt.figure(figsize=(10,5))
plt.style.use('dark_background')
x = np.array([1,2,3,4])*2
plt.bar(x-0.25,[10,20,30,40],width=0.5,tick_label=['Gold','Silver','Platinum','Copper'],label='This year')
plt.bar(x+0.25,[10,30,10,38],width=0.5,label='Next Year')
plt.legend()
plt.show()
```

#### Pie Charts

```python
plt.figure(figsize=(10,5))
plt.style.use('seaborn')
labels = ['Maths','English','Physics','Chemistry']
weight = [10,20,30,40]
colors = ['green','blue','pink','yellow']
plt.pie(weight,labels=labels,explode=[0.01,0.01,0.1,0.01],shadow=True,autopct='%1.1f%%',colors=colors)
plt.legend(loc='best')
plt.show()
```

#### Histograms
-> Here we have studied about the Standard Normal Distribution 


```python
X_sn = np.random.randn(100)
sigma = 5
mean = 70
X = X_sn*sigma + mean
Y = X_sn*4 +60
print(X)
```

```python
#bell shaped curve
plt.figure(figsize=(10,10))
plt.hist(X,label='Marks')
plt.hist(Y,label='Marks1',alpha=0.9)
plt.title('Marks Distribution')
plt.ylabel('Frequency of no of students securing this marks')
plt.xlabel('Marks')
plt.legend()
plt.show()
```
